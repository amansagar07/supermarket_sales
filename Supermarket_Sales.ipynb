```python
sales=sales_df.copy()
```

```python
#Covert month into name using lambda functions
def date_to_month(date_value):
    if date_value ==1:
        return 'January'
    elif date_value ==2:
        return 'February'
    elif date_value ==3:
        return 'March'
    elif date_value ==4:
        return 'April'
    elif date_value ==5:
        return 'May'
    elif date_value ==6:
        return 'June'
    elif date_value ==7:
        return 'July'
    elif date_value ==8:
        return 'August'
    elif date_value ==9:
        return 'September'
    elif date_value ==10:
        return 'October'
    elif date_value ==11:
        return 'November'
    elif date_value ==12:
        return 'December'
```

```python
sales['Months'] = sales['Month'].apply(date_to_month)
```

```python
sales = sales.drop(['Month'],axis=1)
```
